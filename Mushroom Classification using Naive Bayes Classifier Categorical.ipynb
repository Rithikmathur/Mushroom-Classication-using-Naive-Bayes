```python
# This problem could be solve through lidestone smoothing.

# Through lidstone_smoothing function we'll get the probabilitites for value of alpha for a category in a column.

def lidstone_smoothing(ColumnName,Category,alpha):
    
    D = dict()
    
    TrainUniqueFeatureValues = training_data[training_data['class'] == Category][ColumnName].unique()
    
    DataUniqueFeatureValues = data[ColumnName].unique()
    
    for FeatureValue in DataUniqueFeatureValues:
        
        if FeatureValue not in TrainUniqueFeatureValues:
            
            D[FeatureValue] = alpha/(training_data[training_data['class'] == Category].shape[0] + (TrainUniqueFeatureValues.shape[0]*alpha))
        
        else:
            D[FeatureValue] = np.count_nonzero(training_data[training_data['class'] == Category][ColumnName] == FeatureValue)/training_data[training_data['class'] == Category].shape[0]
        
    return D
```

```python
# Here we have computed probabilities for every value of every category of every column.

Dict_of_D = {'p':{}, 'e':{}}

for category in ['p','e']:
    
    for ColumnName in column_names[1:]:
        
        Dict_of_D[category][ColumnName] = lidstone_smoothing(ColumnName,category,0.75)
```

```python
# Here we have got the likelihood probabilities.
Dict_of_D
```

```python
# Claculating Prior Probabilities.
prior_p =  training_data[training_data['class'] == 'p'].shape[0]/training_data.shape[0]

prior_e = training_data[training_data['class'] == 'e'].shape[0]/training_data.shape[0]
```

```python
# In mock_test function data of each mushroom will be inserted at a time.
def mock_test(data):
    
    likelihood_p = 1
    
    likelihood_e = 1
    
    for k,v in data.items(): # It will return the list of key,value tuples.
        
        # Multiplying all 22 probabilities as intersection of random variables is
        # the multiplication of probabilitites(Naive Assumption we take)
        # and we have multiplied it with 1 so we could save the first probability.
        likelihood_p = likelihood_p * Dict_of_D['p'][k][v]
        
        likelihood_e = likelihood_e * Dict_of_D['e'][k][v]
        
    posterior_p = likelihood_p * prior_p
    
    posterior_e = likelihood_e * prior_e
    
    if posterior_p > posterior_e:
        
        return 'p'
    
    else:
        
        return 'e'
```

```python
actual_results = np.array(cv_data['class'])

cv_data = np.array(cv_data.iloc[:,1:])

keys = data.columns[1:]

predicted_category = list()

for i in range(0,cv_data.shape[0]):
    
    data_cv = dict(zip(keys,list(cv_data[i])))
    
    predicted_category.append(mock_test(data_cv))
```

```python
predicted_category = np.array(predicted_category)
```

```python
# Claculating Cross Validation Accuracy
accuracy = np.count_nonzero(predicted_category == actual_results)/actual_results.shape[0]
```

```python
actual_results_test = np.array(testing_data['class'])

test_data = np.array(testing_data.iloc[:,1:])

keys = data.columns[1:]

predicted_category_test = list()

for i in range(0,test_data.shape[0]):
    
    data_test = dict(zip(keys,list(test_data[i]))) #It will then convert the data in dictionary.
    
    predicted_category_test.append(mock_test(data_test)) # It is storing the output we are returning from mock test function.
```
