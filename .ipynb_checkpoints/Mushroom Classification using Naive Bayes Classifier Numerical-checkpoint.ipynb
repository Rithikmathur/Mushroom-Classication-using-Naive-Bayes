We have seen that in categorical data that our accuracy has come very low. So, now we will going to convert the categorical data into the numerical data.

```python
import numpy as np
import pandas as pd
import scipy.stats as s
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\rithi\Mushroom Classication using Naive Bayes\mushrooms.csv")
```

```python
data.head()
```

```python
# Here we are going to assign numbers to the categories.

for column in data.columns[1:]:
    
    data[column].replace(to_replace=list(data[column].unique()),value=list(range(0,len(data[column].unique()))),inplace=True)
```

```python
data
```

```python
# We'll perform one hot encoding on every column.

data_array = list()

for column in data.columns[1:]:
    
    data_array.append(np.eye(len(data[column].unique()),len(data[column].unique()))[data[column]])
```

```python
# The above code will give us the one hot encoded matrices of every column(except the class label)
# and here we are concatenating all the matrices.
X = np.concatenate(data_array,axis=1)
```

```python
X.shape
```

```python
# Converting the data in Zero Mean data
X_dash = X - np.mean(X,axis=0)
```

```python
# Now data has become contineous.
X_dash
```

```python
# We'll calculate the covariance matrix of the zero mean data "X_dash".

X_cov = (1/data.shape[0])*np.matmul(X_dash.T,X_dash)
```

```python
X_cov.shape
```

```python
X_cov
```

```python
# It's a square symmetric matrix. We are going to perform Singular Value Decomposition on it.

factored_matrix = np.linalg.svd(X_cov)
```

```python
# Factored matrix will have three matrices Q(eigen vector), lambda(eigen values), Q_Transpose
# and Q will have the orthonormal eigen Vector.
Q = factored_matrix[0]
```

```python
# Selecting the most contributing eigen vector
Q_tilda = Q[:,0:3]
```

```python
# Projecting data into new vector space.
X_new = np.matmul(X_dash, Q_tilda)
```

```python
X_new.shape
```

```python
labels = np.array(data['class']).reshape(data['class'].shape[0],1)
```

```python
new_df = pd.DataFrame(data=X_new)
```

```python
new_df['class'] = labels
```

```python
new_df
```

```python
# Dividing data for training
training_data_len = int(0.7*data.shape[0])

# Dividing data for cross validation
cv_data_len = int(0.2*data.shape[0])

# Dividing data for testing
testing_data_len = int(0.1*data.shape[0])
```

```python
#creating training data with equal amount of both the categories. In 50:50 ratio
poisonous_train = new_df[new_df['class'] == 'p'].iloc[0:training_data_len//2]
non_poisonous_train = new_df[new_df['class'] == 'e'].iloc[0:training_data_len//2]
```

```python
# Concatinating the both the categories to get complete training data.
training_data = pd.concat([poisonous_train,non_poisonous_train])
```

```python
training_data.shape
```

```python
# Remaining data after excluding training data
poisonous_remain = new_df[new_df['class'] == 'p'].iloc[training_data_len//2:]
non_poisonous_remain = new_df[new_df['class'] == 'e'].iloc[training_data_len//2:]
```

```python
remaining_data = pd.concat([poisonous_remain,non_poisonous_remain])
```

```python
remaining_data
```

```python
remaining_data = np.array(remaining_data)
```

```python
# Assigning random indices equal to the remaining data lenght as it is ordered class wise and we want shuffled data.
random_indices = np.random.choice(np.arange(0,2438),size=(2438,),replace=False)
```

```python
remaining_data = remaining_data[random_indices]
```

```python
# Here we a getting shuffled classes.
remaining_data
```

```python
remaining_data_labels = remaining_data[:,3].reshape(remaining_data.shape[0],1)

# Creating DataFrame for remaining data.
remaining_data = pd.DataFrame(data=remaining_data[:,0:3])

remaining_data['class'] = remaining_data_labels
```

```python
remaining_data
```

```python
# Collecting Cross Validation and Testing data.

cv_data = remaining_data.iloc[0:cv_data_len]

testing_data = remaining_data.iloc[cv_data_len:]
```

```python
# Calculating Mean
mu_hat_p = np.array(training_data[training_data['class'] == 'p'].iloc[:,0:3].mean())

# Calculating Covariance
sigma_hat_p = np.array(training_data[training_data['class'] == 'p'].iloc[:,0:3].cov())
```

```python
# cheking the determinent of the covariance matrix. The more positive it is, the more non singular it is.
np.linalg.det(sigma_hat_p)
```

```python
# Calculating Mean
mu_hat_e = np.array(training_data[training_data['class'] == 'e'].iloc[:,0:3].mean())

# Calculating Covariance
sigma_hat_e = np.array(training_data[training_data['class'] == 'e'].iloc[:,0:3].cov())
```

```python
# cheking the determinent of the covariance matrix
np.linalg.det(sigma_hat_e)
```

```python
# Claculating Prior Probabilities
poisonous_prior = training_data[training_data['class'] == 'p'].shape[0]/training_data.shape[0]

non_poisonous_prior = training_data[training_data['class'] == 'e'].shape[0]/training_data.shape[0]
```

```python
# Applying Naive Bayes Algorithm.

def mock_test(data):
    
    inputs = np.array(data.iloc[:,0:3]) # taking data except for class label
    
    posterior_p = s.multivariate_normal.pdf(inputs,mu_hat_p,sigma_hat_p) * poisonous_prior # Likelihood * Prior

    posterior_e = s.multivariate_normal.pdf(inputs,mu_hat_e,sigma_hat_e) * non_poisonous_prior
    
    boolean_mask = posterior_p > posterior_e # checking condition if probability of muchroom belonging to poisonous is higher or
                                             # not than non poisonous.
    
    predicted_category = pd.Series(boolean_mask)
    
    # replacing values, if it is False it belongs to non poisonous, if it is True then it belongs to poisonous mushroom.
    predicted_category.replace(to_replace = [False,True],value=['e','p'],inplace=True)
    
    return np.array(predicted_category)
```

```python
cv_results = mock_test(cv_data)
```

```python
# Predicted Labels on Cross Validation data.
cv_results
```

```python
actual_results_cv = np.array(cv_data['class'])
```

```python
# Ground Truth Labels in Cross Validation data.
actual_results_cv
```

```python
# Claculating Cross Validation Accuracy
cv_accuracy = np.count_nonzero(cv_results == actual_results_cv)/actual_results_cv.shape[0]
```

```python
cv_accuracy
```

```python
testing_results = mock_test(testing_data)
```

```python
# Predicted Labels on Testing data.
testing_results
```

```python
# Ground Truth Labels in Testing data.
actual_results_test = np.array(testing_data['class'])
```

```python
actual_results_test
```

```python
#Calculating Testing Accuracy.
test_accuracy = np.count_nonzero(testing_results == actual_results_test)/actual_results_test.shape[0]
```

```python
test_accuracy
```

Here, After converting the data in numeric values then we saw the hike in accuracy of about 22% on testing data.
